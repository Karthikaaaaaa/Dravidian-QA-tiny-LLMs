```python
import json
import pandas as pd

# Path to your JSON files
file_path_ml = 'indicqa.ml.json'
file_path_ta = 'indicqa.ta.json'

# Read the JSON files
with open(file_path_ml, 'r') as f:
    data1 = json.load(f)

with open(file_path_ta, 'r') as f:
    data2 = json.load(f)

def count_qa_pairs(data):
    count = 0
    for item in data['data']:
        for paragraph in item['paragraphs']:
            count += len(paragraph['qas'])
    return count

qa_count_ml = count_qa_pairs(data1)
qa_count_ta = count_qa_pairs(data2)

print("Number of question-answer pairs in the Malayalam dataset:", qa_count_ml)
print("Number of question-answer pairs in the Tamil dataset:", qa_count_ta)

# Convert JSON data to DataFrames
df_ml = pd.DataFrame(data1)
df_ta = pd.DataFrame(data2)

print("DataFrames created!")
print("Malayalam DataFrame:")
print(df_ml.head())
print("Tamil DataFrame:")
print(df_ta.head())
```

```python
# Function to create a DataFrame with specific columns
def create_dataframe(data):
    records = []
    for entry in data['data']:
        for paragraph in entry['paragraphs']:
            context = paragraph['context']
            for qa in paragraph['qas']:
                id_ = qa['id']
                category = qa['category']
                question = qa['question']
                # Handling the case where there might be multiple answers, we take the first one
                answer = qa['answers'][0]['text'] if qa['answers'] else ''
                records.append({'id': id_, 'category': category, 'context': context, 'question': question, 'answer': answer})
    return pd.DataFrame(records)

df_ml = create_dataframe(df_ml)
df_ta = create_dataframe(df_ta)

print("DataFrames created wi|th specific columns!")
print("Malayalam DataFrame:")
print(df_ml.head())
print("Tamil DataFrame:")
print(df_ta.head())
```

```python
df_ml['answer'] = df_ml['answer'].replace('', 'no answer')

df_ta['answer'] = df_ta['answer'].replace('', 'no answer')

print("Updated Malayalam DataFrame:")
print(df_ml.head())
print("\nUpdated Tamil DataFrame:")
print(df_ta.head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from rouge_score import rouge_scorer
import numpy as np
import re
from nltk.translate.bleu_score import sentence_bleu # Import sentence_bleu


# Normalize function to clean up text
def normalize_text(text):
    # Strip extra spaces, normalize Unicode characters, lowercase
    text = text.strip().lower()
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with a single space
    return text

# BLEU Score
def compute_bleu(reference, generated):
    reference_tokens = reference.split()  # Simple space-based tokenization for now
    generated_tokens = generated.split()
    if not generated_tokens or not reference_tokens:
        return 0
    return sentence_bleu([reference_tokens], generated_tokens)

# ROUGE Score
def compute_rouge(reference, generated):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, generated)
    return scores

# Cosine Similarity
def compute_cosine_similarity(reference, generated):
    vectorizer = TfidfVectorizer().fit_transform([reference, generated])
    vectors = vectorizer.toarray()
    cosine_sim = cosine_similarity([vectors[0]], [vectors[1]])[0][0]
    return cosine_sim

# Exact Match
def compute_exact_match(reference, generated):
    return int(reference.strip().lower() == generated.strip().lower())

# F1 Score
def compute_f1(reference, generated):
    reference_tokens = reference.split()
    generated_tokens = generated.split()
    common = set(reference_tokens) & set(generated_tokens)
    if len(common) == 0:
        return 0
    precision = len(common) / len(generated_tokens)
    recall = len(common) / len(reference_tokens)
    if precision + recall == 0:
        return 0
    f1_score = 2 * (precision * recall) / (precision + recall)
    return f1_score


# Example QA setup for zero-shot learning

question = "1950 ൽ ഇന്ത്യ വിഭജനത്തിനുശേഷം, ഇന്ഡസ് വാലി  പ്രദേശത്ത് വിപുലമായ പഠനങ്ങൾ നടത്തുകയും കൂടുതൽ അവശിഷ്ടങ്ങൾ കണ്ടെത്തുകയും ചെയ്തതാര് ?"
prompt = f"Question: {question}\nAnswer:"

# Assuming `text_generator` is defined and a model is loaded
outputs = text_generator(
    prompt,
    max_new_tokens=256,
    do_sample=True,
    temperature=0.5,
    top_k=20,
    top_p=0.5
)

generated_text = outputs[0]["generated_text"]
generated_answer = generated_text[len(prompt):].strip()

# Reference answer (from your dataset or manual input)
reference_answer = "മോർട്ടിമർ വീലർ"

# Normalize both answers
reference_answer = normalize_text(reference_answer)
generated_answer = normalize_text(generated_answer)

# Print both answers
print("Generated answer (zero-shot):", generated_answer)
print("Reference answer:", reference_answer)

# Calculate metrics
bleu = compute_bleu(reference_answer, generated_answer)
rouge = compute_rouge(reference_answer, generated_answer)
cosine_sim = compute_cosine_similarity(reference_answer, generated_answer)
exact_match = compute_exact_match(reference_answer, generated_answer)
f1 = compute_f1(reference_answer, generated_answer)

# Display evaluation results
print(f"\nEvaluation Metrics:\n")
print(f"BLEU Score: {bleu}")
print(f"ROUGE Scores: {rouge}")
print(f"Cosine Similarity: {cosine_sim}")
print(f"Exact Match: {exact_match}")
print(f"F1 Score: {f1}")
```

```python
# Step 4: Function to retrieve context from Wikipedia/CC100 using ChromaDB
def retrieve_context(query, n_results=6):
    query_embedding = fasttext_embedding(query)
    query_embedding = [query_embedding.tolist()]

    results = collection.query(
        query_embeddings=query_embedding,
        n_results=n_results
    )

    return results['metadatas']
```

```python
# Step 5: Function to format the retrieved contexts into a prompt
def format_prompt(query, retrieved_contexts):
    contexts = retrieved_contexts[0]
    context_text = "\n".join([f"Context {i+1}: {context['text']}" for i, context in enumerate(contexts)])

    prompt = f"Question: {query}\n\n{context_text}\n\nAnswer the question based on the above contexts."
    return prompt
```

```python
from tqdm import tqdm
import numpy as np
import torch
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

import torch

# Define the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")


# Initialize metrics
total_accuracy = 0
total_f1 = 0
total_precision = 0
total_recall = 0
total_samples = 0

# Custom evaluation loop with tqdm for progress
for batch in tqdm(val_dataset, desc="Evaluating"):
    # Move batch data to device
    input_ids = torch.tensor(batch['input_ids']).to(device)
    attention_mask = torch.tensor(batch['attention_mask']).to(device)
    labels = torch.tensor(batch['labels']).to(device)

    inputs = {
        'input_ids': input_ids.unsqueeze(0),
        'attention_mask': attention_mask.unsqueeze(0),
        'labels': labels.unsqueeze(0)
    }
    
    with torch.no_grad():
        outputs = model_with_lora(**inputs)

    # Get logits
    logits = outputs.logits
    predictions = torch.argmax(logits, dim=-1)  # Get class predictions

    # Print shapes for debugging
    print("Logits shape:", logits.shape)  # Should be [1, 512, num_classes]
    print("Labels shape:", labels.shape)    # Should be [512]
    print("Predictions shape:", predictions.shape)  # Should be [1, 512]

    # Flatten the predictions and labels
    predictions = predictions.view(-1).cpu().numpy()  # Shape: [512]
    labels = labels.view(-1).cpu().numpy()            # Shape: [512]

    # Check lengths
    if len(labels) == len(predictions):
        total_accuracy += accuracy_score(labels, predictions)
        total_f1 += f1_score(labels, predictions, average='weighted', zero_division=0)
        total_precision += precision_score(labels, predictions, average='weighted', zero_division=0)
        total_recall += recall_score(labels, predictions, average='weighted', zero_division=0)
        total_samples += 1
    else:
        print(f"Inconsistent sizes: labels {len(labels)} predictions {len(predictions)}")

# Calculate average metrics
if total_samples > 0:
    avg_accuracy = total_accuracy / total_samples
    avg_f1 = total_f1 / total_samples
    avg_precision = total_precision / total_samples
    avg_recall = total_recall / total_samples

    # Print results
    print("Average Accuracy:", avg_accuracy)
    print("Average F1 Score:", avg_f1)
    print("Average Precision:", avg_precision)
    print("Average Recall:", avg_recall)
else:
    print("No valid samples for metrics calculation.")
```
